```python
import tensorflow as tf
from tensorflow import keras

import numpy as np
import matplotlib.pyplot as plt
import os
import random
import pickle
def build_model(lossfunction):
    if lossfunction == 'mean_squared_error':
        model = keras.Sequential([
        keras.layers.Flatten(input_shape=(48,48,3)),
        keras.layers.Dense(256, activation=tf.nn.relu),
        keras.layers.Dense(128, activation=tf.nn.relu),
        keras.layers.Dense(1, activation="linear")
        ])


    elif lossfunction == 'sparse_categorical_crossentropy':
        model = keras.Sequential([
        keras.layers.Flatten(input_shape=(48,48,3)),
        keras.layers.Dense(256, activation=tf.nn.relu),
        keras.layers.Dense(128, activation=tf.nn.relu),
        keras.layers.Dense(2, activation=tf.nn.softmax)
        ])
    return model

if __name__ == '__main__':
    data = pickle.load(open("dogs_cats.pkl","rb"))
    X_train, Y_train, X_test, Y_test = data["X_train"], data["Y_train"], data["X_test"], data["Y_test"]
    X_train = X_train / 255.0
    X_test = X_test / 255.0
    
    #lossfunction = 'mean_squared_error' or 'sparse_categorical_crossentropy'
    lossfunction = 'mean_squared_error'
    
    model = build_model(lossfunction)
    model.compile(optimizer='adam',
                loss=lossfunction,
                metrics=['accuracy']) 
    model.fit(X_train, Y_train, epochs=20)
    reg = model.predict(X_test).squeeze()
    predictions = np.zeros(reg.shape[0])
    test_accuracy = np.zeros(reg.shape[0])
    
    if lossfunction == 'mean_squared_error':
        for i in range(reg.shape[0]):      
            if reg[i] > 0.5 :
                predictions[i] = 1
            else:
                pass
            if predictions[i] == Y_test[i] :
                test_accuracy[i] = 1
            else:
                pass
       
    elif lossfunction == 'sparse_categorical_crossentropy':  
        for i in range(reg.shape[0]):      
            predictions[i] = np.argmax(reg[i,:])

            if predictions[i] == Y_test[i]:
                test_accuracy[i] = 1
            else:
                pass
            
    print("test_accuracy:",np.mean(test_accuracy))
    print(X_train.shape)
    print(Y_train.shape)
    print(X_test.shape)
    print(Y_test.shape)
```
